doc du lieu va sua tieu de

```python
import pandas as pd

# Đọc dữ liệu và chỉ định dòng đầu tiên làm tiêu đề
url = "https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv"
df = pd.read_csv(url, header=0)  # Sử dụng dòng đầu tiên làm tiêu đề (index 0)

# Kiểm tra 10 dòng đầu tiên
print("10 dòng đầu tiên:")
print(df.head(10))
```

xu ly du lieu trong cot dan toc

```python
# Kiểm tra số lượng dữ liệu thiếu trong cột DT
print("Dữ liệu thiếu trong cột DT:", df['DT'].isnull().sum())

# Điền dữ liệu thiếu bằng 0 trong cột DT
df['DT'] = df['DT'].fillna(0)

# Kiểm tra lại tần số và tần suất
freq_DT = df['DT'].value_counts(dropna=False)
ratio_DT = df['DT'].value_counts(normalize=True, dropna=False)

print("Tần số của DT:\n", freq_DT)
print("Tần suất của DT:\n", ratio_DT)
```

xu ly du lieu thieu trong cac cot diem

```python
# Liệt kê các cột điểm
score_cols = [col for col in df.columns if col.startswith(('T','L','H','S','V','X','D','N','DH'))]

# Điền dữ liệu thiếu cho các cột điểm bằng giá trị trung bình
for col in score_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mean())
```

tao cac bien trung binh mon

```python
# Tạo các biến TBM1, TBM2, TBM3
df['TBM1'] = (df['T1'] + df['L1'] + df['H1'] + df['S1'] + df['V1'] + df['X1'] + df['D1'] + df['N1']) / 8
df['TBM2'] = (df['T2'] + df['L2'] + df['H2'] + df['S2'] + df['V2'] + df['X2'] + df['D2'] + df['N2']) / 8
df['TBM3'] = (df['T6'] + df['L6'] + df['H6'] + df['S6'] + df['V6'] + df['X6'] + df['D6'] + df['N6']) / 8
```

xep loai hoc luc

```python
def classify(x):
    if x < 5:
        return 'Y'
    elif x < 6.5:
        return 'TB'
    elif x < 8:
        return 'K'
    elif x < 9:
        return 'G'
    else:
        return 'XS'

df['XL1'] = df['TBM1'].apply(classify)
df['XL2'] = df['TBM2'].apply(classify)
df['XL3'] = df['TBM3'].apply(classify)
```

chuyen diem trung binh thanh thang diem 10 sang thang diem 4

```python
def min_max_scale(x):
    return (x - 0) / (10 - 0) * 4

df['US_TBM1'] = df['TBM1'].apply(min_max_scale)
df['US_TBM2'] = df['TBM2'].apply(min_max_scale)
df['US_TBM3'] = df['TBM3'].apply(min_max_scale)
```

Tao bien ket qua xet tuyen


```python
def kqxt(row):
    if row['KT'] in ['A', 'A1']:
        score = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    elif row['KT'] == 'B':
        score = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    else:
        score = (row['DH1'] + row['DH2'] + row['DH3']) / 3

    return 1 if score >= 5 else 0

df['KQXT'] = df.apply(kqxt, axis=1)
```

```python
import matplotlib.pyplot as plt

# Biểu đồ tần số (Bar Chart)
df['GT'].value_counts().plot(kind='bar', title='Biểu đồ tần số cho GT')
plt.show()

# Biểu đồ tần suất (Pie Chart)
df['GT'].value_counts(normalize=True).plot(kind='pie', autopct='%1.1f%%', title='Biểu đồ tần suất cho GT')
plt.show()

# Biểu đồ tích lũy tần suất (Cumulative Frequency)
df['GT'].value_counts().sort_index().cumsum().plot(kind='line', title='Biểu đồ tích lũy tần suất cho GT')
plt.show()
```
