```python

import pandas as pd

df = pd.read_csv('/content/sample_data/titanic_disaster.csv')

print("=== TRƯỚC KHI XỬ LÝ ===")
print("Số lượng missing values:")
print(df.isnull().sum())

df_clean = df.copy()

cabin_missing = df_clean['Cabin'].isnull().sum()
cabin_missing_ratio = (cabin_missing / len(df_clean)) * 100
print(f"\nCột Cabin thiếu {cabin_missing}/{len(df_clean)} ({cabin_missing_ratio:.2f}%)")
print("--> Xóa cột Cabin")
df_clean.drop('Cabin', axis=1, inplace=True);

print("\n--- XỬ LÝ CỘT AGE ---")

age_median = df_clean.groupby(['Pclass', 'Sex'])['Age'].median()
print("Median Age theo nhóm Pclass và Sex:")
print(age_median)

def fill_age(row):
    if pd.isnull(row['Age']):
        return age_median[(row['Pclass'], row['Sex'])]
    return row['Age']

df_clean['Age'] = df_clean.apply(fill_age, axis=1)
print(f"Sau khi điền, Age còn thiếu: {df_clean['Age'].isnull().sum()}")

print("\n--- XỬ LÝ CỘT EMBARKED ---")

print("Giá trị hiện tại:")
print(df_clean['Embarked'].value_counts())

embarked_mode = df_clean['Embarked'].mode()[0]
print(f"Mode (giá trị phổ biến nhất): '{embarked_mode}'")

df_clean['Embarked'].fillna(embarked_mode, inplace=True)
print(f"Sau khi điền, Embarked còn thiếu: {df_clean['Embarked'].isnull().sum()}")

print("\n=== SAU KHI XỬ LÝ ===")
print("Số lượng missing values còn lại:")
print(df_clean.isnull().sum())
df_clean.to_csv('titanic_step2_cleaned.csv', index=False)
print("\nĐã lưu file: titanic_step2_cleaned.csv")
```

```python

import pandas as pd
import numpy as np

df = pd.read_csv('titanic_step2_cleaned.csv')

print("=== DỮ LIỆU HIỆN TẠI ===")
print(f"Shape: {df.shape}")
print(df.head())

print("\n=== XỬ LÝ CỘT NAME ===")

def extract_title(name):
    if pd.isna(name):
        return 'Unknown'
    if 'Mr.' in name:
        return 'Mr'
    elif 'Mrs.' in name:
        return 'Mrs'
    elif 'Miss.' in name:
        return 'Miss'
    elif 'Master.' in name:
        return 'Master'
    elif 'Dr.' in name:
        return 'Dr'
    elif 'Rev.' in name:
        return 'Rev'
    elif 'Col.' in name:
        return 'Col'
    elif 'Major.' in name:
        return 'Major'
    elif 'Lady.' in name or 'Countess' in name:
        return 'Lady'
    else:
        return 'Other'

df['Title'] = df['Name'].apply(extract_title)
print("Các Title được tạo:")
print(df['Title'].value_counts())

print("\n=== PHÁT HIỆN OUTLIERS TRONG AGE ===")

q1_age = df['Age'].quantile(0.25)
q3_age = df['Age'].quantile(0.75)
iqr_age = q3_age - q1_age
lower_age = q1_age - 1.5 * iqr_age
upper_age = q3_age + 1.5 * iqr_age

outliers_age = df[(df['Age'] < lower_age) | (df['Age'] > upper_age)]
print(f"Q1 Age: {q1_age:.2f}")
print(f"Q3 Age: {q3_age:.2f}")
print(f"IQR Age: {iqr_age:.2f}")
print(f"Ngưỡng Age: [{lower_age:.2f}, {upper_age:.2f}]")
print(f"Số lượng Age outliers: {len(outliers_age)}")
print("Các Age outliers:")
print(outliers_age[['Age', 'Pclass', 'Sex']].head())

print("\n=== PHÁT HIỆN OUTLIERS TRONG FARE ===")

q1_fare = df['Fare'].quantile(0.25)
q3_fare = df['Fare'].quantile(0.75)
iqr_fare = q3_fare - q1_fare
lower_fare = q1_fare - 1.5 * iqr_fare
upper_fare = q3_fare + 1.5 * iqr_fare

outliers_fare = df[(df['Fare'] < lower_fare) | (df['Fare'] > upper_fare)]
print(f"Q1 Fare: {q1_fare:.2f}")
print(f"Q3 Fare: {q3_fare:.2f}")
print(f"IQR Fare: {iqr_fare:.2f}")
print(f"Ngưỡng Fare: [{lower_fare:.2f}, {upper_fare:.2f}]")
print(f"Số lượng Fare outliers: {len(outliers_fare)}")

print("\n=== XỬ LÝ OUTLIERS ===")

print("Cách 1: Giữ nguyên outliers nhưng ghi nhận")

print("\nCách 2: Giới hạn outliers - Có thể áp dụng nếu cần")

print("\n=== TẠO FEATURE MỚI ===")

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print("FamilySize (bao gồm bản thân):")
print(df['FamilySize'].value_counts().sort_index())

df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
print("\nIsAlone (đi một mình):")
print(df['IsAlone'].value_counts())

df.to_csv('titanic_step3_normalized.csv', index=False)
print("\nĐã lưu file: titanic_step3_normalized.csv")
```
